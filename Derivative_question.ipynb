```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data\\04b_Exercise_velocity_acceleration_data_file.dat", names= ['time', 'position'], sep= ' ')
```

```python
plt.scatter(df['time'], df['position'])
plt.xlabel("time")
plt.ylabel("position")
plt.show()      ### ----(random function path)
```

```python
def forward(data):
    derive = np.zeros(len(df[:, 1]))
    ## for equidistant data
    ## h = (data[1, 0] - data[0,0])
    ## velocity[:, -1] = (data[:-1, 1] - data[1:, 1]) / h
    # general
    derive[:-1] = (data[1:, 1] - data[:-1, 1]) / (data[1:, 0] - data[:-1, 0])
    derive[-1] = (data[-1, 1] - data[-2, 1]) / (data[-1, 0] - data[-2, 0])
    
    return np.transpose([data[:, 0], derive])
```

```python
velocityforward = forward(df)
accelerationforward = forward(velocityforward)

plt.subplot(2, 1, 1)
plt.plot(df['time'], velocityforward, 'green')
plt.xlabel("time")
plt.ylabel('velocity')
plt.grid(True)
plt.legend()

plt.subplot(2,1,2)
plt.plot(df['time'], accelerationforward, 'green')
plt.xlabel("time")
plt.ylabel('acceleration')
plt.grid(True)
plt.legend()
```

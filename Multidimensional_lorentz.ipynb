## Lorenz system

We solve the three coupled differential equations

$$
\dot{x} = a(y-x)\\
\dot{y} = x(b-z)-y\\
\dot{z} = xy-cz\\
$$

Here $ x,y,z $ are the coordinates and $ a,b,c $ are coefficients

```python
### x, y, z are time dependent variables
## a, b, c are parameters influencing the behaviour of system

## x represents the convection intensity (horizontal temp variation)
## y corresponds to temperature difference
## z represents to vertical temperature profile

## a known ad prandtl number = (fluid viscosity / thermal diffusivity)
## b known as reyleigh number = (temp gradient driving convection)
## c geometric error = 8/3 (generally)
```

##### Dynamics of System:
1. Fixed points:  For certain parameter values, the system has fixed points that can be stable or unstable. These points represent steady states where time derivative of x, y, z are zero.
2. Chaos: For specific ranges ofa, b, c the Lorenz system demonstrates chaotic behavior, characterized by: Sensitive dependence on initial conditions (butterfly effect). A strange attractorâ€”a fractal structure in phase space that governs the system's trajectory.
3. Bifurcations: As b or other parameters vary, the system undergoes transitions (bifurcations) from stable fixed points to periodic oscillations to chaotic motion.

```python


def lorentz_system(t, r, a, b, c):
    x, y, z = r
    x1 = a*(y - x)
    y1 = x*(b - z) - y
    z1 = x*y - c*z
    return [x1, y1, z1]
```

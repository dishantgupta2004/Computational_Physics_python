#### Rutherford scattering using Gaussian

```python
Z= 79
e = 1.602e-19
E = 7.7e6 * e
epsilon0 = 8.854e-12
a0 = 5.292e-11
sigma = a0 / 100
N = 1000000

def gaussian():
    r = np.sqrt(-2*sigma*sigma*np.log(1-random.random()))
    theta = 2*np.pi*random.random()
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return x, y

## Main Program
count = 0
for i in range(N):
    x, y = gaussian()
    b = np.sqrt(x*x + y*y)
    if b < Z*e*e/(2*np.pi*epsilon0*E):
        count +=1 

print(count, "particles were reflected out of ", N)
```

### Monte-Carlo Integration

```python
from random import random
def f(x):
    return (np.sin(1/x*(2-x)))**2 if x != 0 else 0

# Hit-or-Miss Monte Carlo Method
N = 1000
count = 0
for i in range(N):
    x = 2 * random()
    y = random()
    if y < f(x):
        count += 1

hit_or_miss_integral = 2 * count / N

# Mean Value Monte Carlo Method
mean_value_integral = 0
for i in range(N):
    x = 2 *random()
    mean_value_integral += f(x)

mean_value_integral *= 2 / N

# Exact Value (Numerical Integration for Comparison)
from scipy.integrate import quad
exact_integral, _ = quad(lambda x: f(x), 1e-5, 2)

# Errors
hit_or_miss_error = abs(hit_or_miss_integral - exact_integral)
mean_value_error = abs(mean_value_integral - exact_integral)

print(f"Hit-or-Miss Monte Carlo Integral: {hit_or_miss_integral}")
print(f"Mean Value Monte Carlo Integral: {mean_value_integral}")
print(f"Exact Integral: {exact_integral}")
print(f"Hit-or-Miss Error: {hit_or_miss_error}")
print(f"Mean Value Error: {mean_value_error}")
```

- Monte Carlo integration is especially useful when:
1. The integral is high-dimensional, making traditional quadrature methods impractical.
2. The domain is irregular or the integrand is highly non-linear.
3. Analytical solutions are intractable or do not exist.
4. Monte Carlo is not constrained by the dimensionality of the integral, as the number of samples mscales independently of the dimensionality. This makes it powerful for solving problems in many-body physics, quantum mechanics, or statistical physics.

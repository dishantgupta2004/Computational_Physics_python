```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, t):
    return -x**3 + np.sin(t)

a = 0.0   ## start
b = 10.0  ## End 
N = 100  ## Number of steps
h = (b-a) / N   ## Step Size
x = 0.0   ## Initial Condition

tpoints = np.arange(a, b, h)
xpoints = []
for t in tpoints:
    xpoints.append(x)
    x += h * f(x, t)
    
plt.plot(tpoints, xpoints)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```

```python
## Similar ODE Using runga kutta of 2nd Order
tpoints_runga = np.arange(a, b, h)
xpoints_runga2 = []

x = 0.0
for t in tpoints_runga:
    xpoints_runga2.append(x)
    k1 = h * f(x, t)
    k2 = h * f(x+0.5*k1, t+0.5*h)
    x += k2
    
plt.plot(tpoints, xpoints_runga2, label= 'Runge 2nd Order', color= 'green')
plt.plot(tpoints, xpoints, label= 'Euler')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.show()
```

```python
## Similar ODE Using runga kutta of 2nd Order
tpoints_runga = np.arange(a, b, h)
xpoints_runga4 = []

x = 0.0
for t in tpoints_runga:
    xpoints_runga4.append(x)
    k1 = h*f(x,t)
    k2 = h*f(x+0.5*k1,t+0.5*h)
    k3 = h*f(x+0.5*k2,t+0.5*h)
    k4 = h*f(x+k3,t+h)
    x += (k1+2*k2+2*k3+k4)/6
    
plt.plot(tpoints, xpoints_runga2, label= 'Runge 2nd Order', color= 'green')
plt.plot(tpoints, xpoints, label= 'Euler')
plt.plot(tpoints, xpoints_runga4, label= 'Runga 4rd Order')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.show()
```
